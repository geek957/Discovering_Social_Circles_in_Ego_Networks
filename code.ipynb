```python
def calcdist(v1,v2):
    ans=0
    for i in range(len(v1)):
        ans+=v1[i]*v2[i]
    return ans
```

```python
def max_circles():
    maxdist=-1;
    combined=[1,2]
    for i in range(len(features)):
        for j in range(i+1,len(features)):
#             print i,j
            if(vis[i]==1 or vis[j]==1):
                continue
            if(calcdist(features[i],features[j])>maxdist):
                maxdist=calcdist(features[i],features[j]);
                combined[0]=i;
                combined[1]=j;
    return combined
            
```
